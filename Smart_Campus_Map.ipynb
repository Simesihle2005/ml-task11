List of locations

```python
locations = [
    "Admin", "Engineering", "Science", "Library", "ICT",
    "ResA", "ResB", "Cafeteria", "Sports", "MainEntrance"
]
```

Directed edges (routes)

```python
edges = {
    "Admin": ["Library"],
    "Library": ["ICT", "Sports"],
    "ICT": ["Engineering"],
    "Engineering": ["Science"],
    "Science": ["Cafeteria"],
    "Cafeteria": ["ResA"],
    "ResA": ["ResB"],
    "ResB": ["MainEntrance"],
    "MainEntrance": ["Admin"]
}
```

adjacency list

```python
print("Adjacency List:\n")
for loc in locations:
    connected = edges.get(loc, [])
    print(f"{loc} → {connected}")
```

Adjacency matxix

```python
print("\nAdjacency Matrix:\n")
```

Initialize matrix with zeros

```python
matrix = [[0 for _ in locations] for _ in locations]
```

Create name-index mapping

```python
index_map = {name: i for i, name in enumerate(locations)}
```

Fill the matrix

```python
for src, destinations in edges.items():
    for dst in destinations:
        i = index_map[src]
        j = index_map[dst]
        matrix[i][j] = 1
```

Print header

```python
print("     " + "  ".join(f"{loc[:3]}" for loc in locations))
for i, row in enumerate(matrix):
    print(f"{locations[i][:3]}  {row}")
```

Directed graph

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from([
    ("Admin", "Library"),
    ("Library", "ICT"),
    ("Library", "Sports"),
    ("ICT", "Engineering"),
    ("Engineering", "Science"),
    ("Science", "Cafeteria"),
    ("Cafeteria", "ResA"),
    ("ResA", "ResB"),
    ("ResB", "MainEntrance"),
    ("MainEntrance", "Admin")
])

plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=2000, arrows=True, font_size=10)
plt.title("Smart Campus Shuttle System - Directed Graph")
plt.show()
```
